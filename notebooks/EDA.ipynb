# Exploratory Data Analysis (EDA)

This notebook is dedicated to performing exploratory data analysis on the movie dataset. The goal is to understand the data better, visualize key aspects, and derive insights that can inform further analysis or model development.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style='whitegrid')
```

```python
import pandas as pd

# Load datasets
movies_df = pd.read_csv('../data/movies_dataset.csv')
credits_df = pd.read_csv('../data/credits.csv')

# Display the first few rows of the movies dataset
print(movies_df.head())
```

```python
# Display the first few rows of the credits dataset
credits_df.head()
```

```python
# Check for missing values in the movies dataset
movies_df.isnull().sum()
```

```python
# Visualize the distribution of movie ratings
plt.figure(figsize=(10, 6))
sns.histplot(movies_df['vote_average'], bins=20, kde=True)
plt.title('Distribution of Movie Ratings')
plt.xlabel('Vote Average')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze the relationship between budget and revenue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=movies_df, x='budget', y='revenue')
plt.title('Budget vs Revenue')
plt.xlabel('Budget')
plt.ylabel('Revenue')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
# Importar bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset procesado
processed_df = pd.read_csv('../data/movies_processed.csv', low_memory=False)

# Mostrar información básica del dataset
print("Información del dataset:")
print(processed_df.info())

print("\nPrimeras filas del dataset:")
display(processed_df.head())
```

```python
# Resumen estadístico
print("Resumen estadístico de variables numéricas:")
display(processed_df.describe())

# Conteo de valores únicos para variables categóricas
print("\nConteo de valores únicos para variables categóricas:")
for col in processed_df.select_dtypes(include='object').columns:
    print(f"{col}: {processed_df[col].nunique()} valores únicos")
```

```python
# Distribución de revenue
plt.figure(figsize=(12, 6))
sns.histplot(processed_df['revenue'], bins=50, kde=True)
plt.title('Distribución de Revenue')
plt.xlabel('Revenue')
plt.ylabel('Frecuencia')
plt.show()

# Distribución de budget
plt.figure(figsize=(12, 6))
sns.histplot(processed_df['budget'], bins=50, kde=True)
plt.title('Distribución de Budget')
plt.xlabel('Budget')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Distribución de return
plt.figure(figsize=(12, 6))
sns.histplot(processed_df['return'], bins=50, kde=True)
plt.title('Distribución de Return (Retorno de Inversión)')
plt.xlabel('Return')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Distribución de release_year
plt.figure(figsize=(12, 6))
sns.histplot(processed_df['release_year'], bins=50, kde=True)
plt.title('Distribución de Años de Estreno')
plt.xlabel('Año de Estreno')
plt.ylabel('Frecuencia')
plt.show()
```

### Analisis de Correlaciones

```python
# Matriz de correlación
numeric_df = processed_df.select_dtypes(include=[np.number])
correlation_matrix = numeric_df.corr()

# Mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```
